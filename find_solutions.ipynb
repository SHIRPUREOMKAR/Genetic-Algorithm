```python
import random
```

```python
def value(x, y, z):
  return 3*x**4 + 3*y**6 + 3*z - 679
```

```python
def score(x, y, z):
  if (value(x, y, z) == 0):
    return 999999
  else:
    return abs(1/value(x, y, z))
```

```python
gens = []

for s in range(1000):
  gens.append( (random.uniform(0,1000), random.uniform(0,1000), random.uniform(0,1000)) )
```

```python
value(37.8385, 96.6348, 10.7358)
```

```python
for i in range(10000):
  ranked = []
  for s in gens:
    ranked.append( (score(s[0], s[1], s[2]), s) )
  ranked.sort()
  ranked.reverse()
  print(f"gen {i} score with {ranked[0]} with values {ranked[0][1][0]}, {ranked[0][1][1]}, {ranked[0][1][2]}")
  selected = ranked[:100]
  elem = []
  if ranked[0][0] > 9999  :
    print(value(ranked[0][1][0], ranked[0][1][1], ranked[0][1][2]))

    break
  for s in selected:
    elem.append(s[1][0])
    elem.append(s[1][1])
    elem.append(s[1][2])

  newgen = []
  for i in range(1000):
    newgen.append( (random.choice(elem)* random.uniform(0.99, 1.01), random.choice(elem)* random.uniform(0.99, 1.01), random.choice(elem)* random.uniform(0.99, 1.01)))
  gens = newgen
```
